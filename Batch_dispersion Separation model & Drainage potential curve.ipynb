```python
import math
import numpy as np
import matplotlib.pyplot as plt
import tkinter as tk
from PIL import ImageTk, Image
import os
from tkinter import messagebox, Scale
from tkinter import ttk, Scale
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

# Create the main tkinter window
root = tk.Tk()
root.title('Plotter GUI')

# Create labels and entry fields for inputs
water_density_label = tk.Label(root, text='Water Density (kg/m^3):')
water_density_label.grid(row=0, column=0, padx=10, pady=10)

water_density_value = tk.StringVar(value='1020.7')
water_density_entry = tk.Entry(root, textvariable=water_density_value)
water_density_entry.grid(row=0, column=1, padx=10, pady=10)

water_viscosity_label = tk.Label(root, text='Water Viscosity (Pa.s):')
water_viscosity_label.grid(row=1, column=0, padx=10, pady=10)

water_viscosity_value = tk.StringVar(value='0.001')
water_viscosity_entry = tk.Entry(root, textvariable=water_viscosity_value)
water_viscosity_entry.grid(row=1, column=1, padx=10, pady=10)

water_fraction_label = tk.Label(root, text='Water Fraction (-):')
water_fraction_label.grid(row=2, column=0, padx=10, pady=10)

water_fraction_value = tk.StringVar(value='0.5')
water_fraction_entry = tk.Entry(root, textvariable=water_fraction_value)
water_fraction_entry.grid(row=2, column=1, padx=10, pady=10)

oil_density_label = tk.Label(root, text='Oil Density (kg/m^3):')
oil_density_label.grid(row=3, column=0, padx=10, pady=10)

oil_density_value = tk.StringVar(value='787.45')
oil_density_entry = tk.Entry(root, textvariable=oil_density_value)
oil_density_entry.grid(row=3, column=1, padx=10, pady=10)

oil_viscosity_label = tk.Label(root, text='Oil Viscosity (Pa.s):')
oil_viscosity_label.grid(row=4, column=0, padx=10, pady=10)

oil_viscosity_value = tk.StringVar(value='0.0013')
oil_viscosity_entry = tk.Entry(root, textvariable=oil_viscosity_value)
oil_viscosity_entry.grid(row=4, column=1, padx=10, pady=10)

droplet_diameter_label = tk.Label(root, text='Droplet Diameter (m):')
droplet_diameter_label.grid(row=5, column=0, padx=10, pady=10)

droplet_diameter_value = tk.StringVar(value='0.000265')
droplet_diameter_entry = tk.Entry(root, textvariable=droplet_diameter_value)
droplet_diameter_entry.grid(row=5, column=1, padx=10, pady=10)

inversion_point_label = tk.Label(root, text='Inversion Point:')
inversion_point_label.grid(row=6, column=0, padx=10, pady=10)

inversion_point_value = tk.StringVar(value='0.33')
inversion_point_entry = tk.Entry(root, textvariable=inversion_point_value)
inversion_point_entry.grid(row=6, column=1, padx=10, pady=10)

flow_rate_lpm_label = tk.Label(root, text='Flow Rate (LPM):')
flow_rate_lpm_label.grid(row=7, column=0, padx=10, pady=10)

flow_rate_lpm_value = tk.StringVar(value='300')
flow_rate_lpm_entry = tk.Entry(root, textvariable=flow_rate_lpm_value)
flow_rate_lpm_entry.grid(row=7, column=1, padx=10, pady=10)

pipe_diameter_m_label = tk.Label(root, text='Pipe Diameter (m):')
pipe_diameter_m_label.grid(row=8, column=0, padx=10, pady=10)

pipe_diameter_m_value = tk.StringVar(value='0.1524')
pipe_diameter_m_entry = tk.Entry(root, textvariable=pipe_diameter_m_value)
pipe_diameter_m_entry.grid(row=8, column=1, padx=10, pady=10)

number_branches_label = tk.Label(root, text='Number of Branches:')
number_branches_label.grid(row=9, column=0, padx=10, pady=10)

number_branches_value = tk.StringVar(value='2')
number_branches_entry = tk.Entry(root, textvariable=number_branches_value)
number_branches_entry.grid(row=9, column=1, padx=10, pady=10)

interfacial_tension_label = tk.Label(root, text='Interfacial Tension (Nm/m):')
interfacial_tension_label.grid(row=10, column=0, padx=10, pady=10)

interfacial_tension_value = tk.StringVar(value='28.8')
interfacial_tension_entry = tk.Entry(root, textvariable=interfacial_tension_value)
interfacial_tension_entry.grid(row=10, column=1, padx=10, pady=10)

section_length_label = tk.Label(root, text='Tapping Point Location (m):')
section_length_label.grid(row=7, column=2, padx=10, pady=10)

section_length_value = tk.StringVar(value='4')
section_length_entry = tk.Entry(root, textvariable=section_length_value)
section_length_entry.grid(row=7, column=3, padx=10, pady=10)

# Add sliders for asymmetry_parameter and oil_water_contaminationNR
asymmetry_parameter_label = tk.Label(root, text='Asymmetry Parameter Initial Guess:')
asymmetry_parameter_label.grid(row=5, column=2, padx=10, pady=10)

asymmetry_parameter_value = tk.DoubleVar(value=0.003)
asymmetry_parameter_slider = Scale(root, from_=0, to=0.1, resolution=0.001, orient=tk.HORIZONTAL, variable=asymmetry_parameter_value)
asymmetry_parameter_slider.grid(row=5, column=3, padx=10, pady=10)

oil_water_contaminationNR_label = tk.Label(root, text='WiO layer Contamination Initial Guess:')
oil_water_contaminationNR_label.grid(row=6, column=2, padx=10, pady=10)

oil_water_contaminationNR_value = tk.DoubleVar(value=0.2)
oil_water_contaminationNR_slider = Scale(root, from_=0, to=1, resolution=0.01, orient=tk.HORIZONTAL, variable=oil_water_contaminationNR_value)
oil_water_contaminationNR_slider.grid(row=6, column=3, padx=10, pady=10)


# Create a parent frame for the plots
plot_parent_frame = ttk.Frame(root)
plot_parent_frame.grid(row=5, column=5, rowspan=6, columnspan = 4, padx=10, pady=10, sticky='nsew')

# Create frames for individual plots
plot_frame1 = ttk.Frame(plot_parent_frame)
plot_frame1.pack(side=tk.LEFT, fill=tk.BOTH, expand=1)

plot_frame2 = ttk.Frame(plot_parent_frame)
plot_frame2.pack(side=tk.LEFT, fill=tk.BOTH, expand=1)

# Create a parent frame for the plots
plot_parent_frame2 = ttk.Frame(root)
plot_parent_frame2.grid(row=11, column=0, rowspan=10, columnspan = 9, padx=10, pady=10, sticky='nsew')

# Create frames for individual plots
plot_frame3 = ttk.Frame(plot_parent_frame2)
plot_frame3.pack(side=tk.LEFT, fill=tk.BOTH, expand=1)

plot_frame4 = ttk.Frame(plot_parent_frame2)
plot_frame4.pack(side=tk.LEFT, fill=tk.BOTH, expand=1)

plot_frame5 = ttk.Frame(plot_parent_frame2)
plot_frame5.pack(side=tk.LEFT, fill=tk.BOTH, expand=1)

# Function to handle button click
def on_button_click():
    
        # Clear previous plots
    for widget in plot_frame1.winfo_children():
        widget.destroy()
    for widget in plot_frame2.winfo_children():
        widget.destroy()
    for widget in plot_frame3.winfo_children():
        widget.destroy()
    for widget in plot_frame4.winfo_children():
        widget.destroy()
    for widget in plot_frame5.winfo_children():
        widget.destroy()
        
        
    # Input values
    water_fraction = float(water_fraction_entry.get())
    interfacial_tension = float(interfacial_tension_entry.get())
    asymmetry_parameter = float(asymmetry_parameter_value.get())
    oil_water_contaminationNR = float(oil_water_contaminationNR_value.get())
    section_length = float(section_length_value.get())
    flow_rate_lpm = float(flow_rate_lpm_entry.get())
    pipe_diameter_m = float(pipe_diameter_m_entry.get())
    number_branches = int(number_branches_entry.get())
    water_density = float(water_density_entry.get())
    water_viscosity = float(water_viscosity_entry.get())
    oil_density = float(oil_density_entry.get())
    oil_viscosity = float(oil_viscosity_entry.get())
    droplet_diameter = float(droplet_diameter_entry.get())
    inversion_point = float(inversion_point_entry.get())

    # Calculate flow parameters
    flow_paramsrams = calculate_flow_parameters(flow_rate_lpm, pipe_diameter_m, number_branches)
    mixture_velocity_mps = flow_paramsrams['mixture_velocity_mps']

    # Calculate slip velocity
    slip_velocity, continuous_viscosity = calculate_slip_velocity(
        water_density, water_viscosity, water_fraction, oil_density, oil_viscosity, droplet_diameter, inversion_point)

    # Calculate water layer thickness along the pipe
    water_layer_thickness_along_pipe, water_layer_thickness_final = calculate_water_layer_thickness(
        slip_velocity, pipe_diameter_m, mixture_velocity_mps, water_fraction)
    
    print(water_layer_thickness_along_pipe)

    # Define the Newton-Raphson method parameters
    tolerance = 1e-6
    max_iterations = 100

    # Store history for analysis (if needed)
    error_history = []

    # Newton-Raphson optimization loop
    for iteration in range(1, max_iterations + 1):
        # Calculate oil layer parameters with current asymmetry_parameter
        oil_layer_parameters = calculate_oil_layer_parameters(water_layer_thickness_along_pipe, water_density, oil_density, interfacial_tension, continuous_viscosity, asymmetry_parameter, mixture_velocity_mps)

        # Calculate error
        total_oil_thickness = pipe_diameter_m - water_layer_thickness_final
        error = (((oil_layer_parameters[-1, 2] ** 2) - (total_oil_thickness ** 2)) / (total_oil_thickness ** 2))

        # Calculate derivative of error with respect to asymmetry_parameter (approximation)
        delta = 1e-6
        asymmetry_parameter_delta = asymmetry_parameter + delta
        oil_layer_parameters_delta = calculate_oil_layer_parameters(water_layer_thickness_along_pipe, water_density, oil_density, interfacial_tension, continuous_viscosity, asymmetry_parameter_delta, mixture_velocity_mps)
        error_delta = (((oil_layer_parameters_delta[-1, 2] ** 2) - (total_oil_thickness ** 2)) / (total_oil_thickness ** 2))
        derivative_error = (error_delta - error) / delta

        # Update asymmetry_parameter using Newton-Raphson method
        asymmetry_parameter -= error / derivative_error

        # Store error history for analysis (optional)
        error_history.append(error)

        # Print current iteration information (optional)
        #print(f"Iteration {iteration}: Asymmetry Parameter = {asymmetry_parameter}, Error = {error}")

        # Check for convergence
        if abs(error) < tolerance:
            print(f"Converged in {iteration} iterations.")
            break
    else:
        print("Warning: Maximum iterations reached without convergence for Asymmetry Parameter.")



    # Final calculation of oil layer parameters with optimized asymmetry_parameter
    oil_layer_parameters = calculate_oil_layer_parameters(water_layer_thickness_along_pipe, water_density, oil_density, interfacial_tension, continuous_viscosity, asymmetry_parameter, mixture_velocity_mps)



        # Get the section length from the user
        #section_length = float(input(f"Enter the horizontal pipe section length (0 to {water_layer_thickness_along_pipe[-1, 0]}): "))



    # Ensure the input is within the valid range
    if section_length < 0 or section_length > water_layer_thickness_along_pipe[-1, 0]:
        raise ValueError("Section length is out of the valid range.")

    # Interpolate the water and oil layer thickness at the specified section length
    water_layer_thickness_at_section = np.interp(section_length, water_layer_thickness_along_pipe[:, 0], water_layer_thickness_along_pipe[:, 1])
    oil_layer_thickness_at_section = np.interp(section_length, oil_layer_parameters[:, 0], oil_layer_parameters[:, 2])

    # Calculate the emulsion thickness
    emulsion_thickness_at_section = pipe_diameter_m - (water_layer_thickness_at_section + oil_layer_thickness_at_section)

    # Report the values
    print(f"At section length {section_length} m:")
    print(f"Water Layer Thickness (m): {water_layer_thickness_at_section}")
    print(f"Oil Layer Thickness (m): {oil_layer_thickness_at_section}")
    print(f"Emulsion Thickness (m): {emulsion_thickness_at_section}")

    print(f"Water Layer fraction (m): {water_layer_thickness_at_section / pipe_diameter_m}")
    print(f"Oil Layer frcation (m): {oil_layer_thickness_at_section / pipe_diameter_m}")
    print(f"Emulsion fraction (m): {emulsion_thickness_at_section / pipe_diameter_m}")

    # Third plot
    fig3, ax3 = plt.subplots(figsize=(8, 4))
    ax3.plot(water_layer_thickness_along_pipe[:, 0], water_layer_thickness_along_pipe[:, 1], label='Water Layer Thickness')
    ax3.plot(oil_layer_parameters[:, 0], pipe_diameter_m - oil_layer_parameters[:, 2], label='Oil Layer Thickness')
    ax3.set_xlabel('Pipe Length (m)')
    ax3.set_ylabel('Pipe Diameter (m)')
    ax3.set_ylim(0, pipe_diameter_m)
    ax3.set_xlim(0, water_layer_thickness_along_pipe[-1, 0])
    ax3.set_title('Layers Thickness Profile along the Pipe Length')
    ax3.legend()
    ax3.grid(True)
    canvas3 = FigureCanvasTkAgg(fig3, master=plot_frame3)
    canvas3.draw()
    canvas3.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)

    # Constants and parameters
    D_in = 1  # [m] Inner diameter of the pipe
    initial_water_cut = water_fraction  # Initial cumulative water fraction desired
    emulsion_thickness = emulsion_thickness_at_section / pipe_diameter_m # 20% of the pipe diameter
    water_contamination = 1  #% water in the water layer
    #--------------------------------------------------------------------------
    # Define the Newton-Raphson method parameters
    tolerance3 = 1e-6
    max_iterations2 = 100

    # Store history for analysis (if needed)
    error3_history = []

    # Newton-Raphson optimization loop
    for iteration2 in range(1, max_iterations2 + 1):
        # Calculate oil layer parameters with current asymmetry_parameter
        # Find layer thicknesses
        oil_thicknessNR, water_thicknessNR, cumulative_water_fractionNR = find_layer_thicknesses(initial_water_cut, emulsion_thickness, oil_water_contaminationNR, water_contamination, D_in)

        # Calculate error
        oil_fractionNR = oil_thicknessNR / D_in
        error3 = (((oil_fractionNR ** 2) - ((oil_layer_thickness_at_section / pipe_diameter_m) ** 2)) / ((oil_layer_thickness_at_section / pipe_diameter_m) ** 2))

        # Calculate derivative of error with respect to asymmetry_parameter (approximation)
        delta3 = 1e-3
        oil_water_contaminationNR_delta = oil_water_contaminationNR + delta3
        oil_thicknessNR_delta, water_thicknessNR_delta, cumulative_water_fractionNR_delta = find_layer_thicknesses(initial_water_cut, emulsion_thickness, oil_water_contaminationNR_delta, water_contamination, D_in)
        error3_delta = (((oil_thicknessNR_delta ** 2) - ((oil_layer_thickness_at_section / pipe_diameter_m) ** 2)) / ((oil_layer_thickness_at_section / pipe_diameter_m) ** 2))
        derivative_error3 = (error3_delta - error3) / delta3

        # Update asymmetry_parameter using Newton-Raphson method
        oil_water_contaminationNR -= error3 / derivative_error3

        # Store error history for analysis (optional)
        error3_history.append(error3)

        # Print current iteration information (optional)
        #print(f"Iteration {iteration2}: Asymmetry Parameter = {oil_water_contaminationNR}, Error = {error3}")

        # Check for convergence
        if abs(error3) < tolerance3:
            print(f"Converged in {iteration2} iterations.")
            break
    else:
        print("Warning: Maximum iterations reached without convergence for WiO layer contamination.")
    #--------------------------------------------------------------------------

    oil_water_contamination = oil_water_contaminationNR  #% water in the oil layer

    # Discretize the pipe diameter into segments
    y = np.linspace(0, D_in, 1000)

    # Calculate the widths at each height
    widths = width_func(y, D_in)

    # Find layer thicknesses
    oil_thickness, water_thickness, cumulative_water_fraction = find_layer_thicknesses(initial_water_cut, emulsion_thickness, oil_water_contamination, water_contamination, D_in)

    # Calculate the water fraction profile
    water_fraction = calculate_water_fraction(y, oil_thickness, emulsion_thickness, oil_water_contamination, water_contamination)

    # Calculate areas, water fraction and WT
    cumulative_areas, cumulative_water_areas, WT = calculate_areas_and_water_fraction(D_in, y, widths, water_fraction, initial_water_cut)

    # Plotting the error vs. iteration (optional)
    #plt.figure(figsize=(2, 4))
    #plt.plot(range(1, iteration + 1), error_history, marker='o', linestyle='-', color='b', markersize=5)
    #plt.xlabel('Iteration')
    #plt.ylabel('Error')
    #plt.title('asymmetry Parameter Iteration')
    #plt.grid(True)
    #plt.tight_layout()
    #plt.show()
    

    
    # First plot
    fig1, ax1 = plt.subplots(figsize=(4, 4))
    ax1.plot(range(1, iteration + 1), error_history, marker='o', linestyle='-', color='b', markersize=5)
    ax1.set_xlabel('Iteration')
    ax1.set_ylabel('Error')
    ax1.set_title('Asymmetry Parameter Iteration')
    ax1.grid(True)
    canvas1 = FigureCanvasTkAgg(fig1, master=plot_frame1)
    canvas1.draw()
    canvas1.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)
    
    # Second plot
    fig2, ax2 = plt.subplots(figsize=(4, 4))
    ax2.plot(range(1, iteration2 + 1), error3_history, marker='o', linestyle='-', color='r', markersize=5)
    ax2.set_xlabel('Iteration')
    ax2.set_ylabel('Error')
    ax2.set_title('Water Contamination in Oil Layer Iteration')
    ax2.grid(True)
    canvas2 = FigureCanvasTkAgg(fig2, master=plot_frame2)
    canvas2.draw()
    canvas2.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)
    

    
    # Fourth plot
    fig4, ax4 = plt.subplots(figsize=(2, 4))
    ax4.plot(water_fraction, y, label='Water Fraction')
    ax4.plot(cumulative_water_areas, y, label='Cumulative Water Fraction')
    ax4.plot(widths, y, label='Pipe Widths', color='gray')
    ax4.axvline(initial_water_cut, color='r', linestyle='--', label=f'Total Water Cut = {initial_water_cut:.2f}')
    ax4.set_ylabel('Pipe Diameter [m]')
    ax4.set_xlabel('Water Fraction (WC) [-]')
    ax4.set_title('Water Fraction Change from Bottom to Top of the Pipe')
    ax4.legend()
    ax4.grid(True)
    ax4.set_ylim(0, D_in)
    canvas4 = FigureCanvasTkAgg(fig4, master=plot_frame4)
    canvas4.draw()
    canvas4.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)
    
    # Fifth plot
    fig5, ax5 = plt.subplots(figsize=(3, 4))
    ax5.plot(water_fraction, WT, label='Drainage Potential Curve')
    ax5.set_ylabel('Water Tapped (WT) [%]')
    ax5.set_xlabel('Water cut tapped stream [%]')
    ax5.set_title('Drainage Potential Curve')
    ax5.legend()
    ax5.grid(True)
    canvas5 = FigureCanvasTkAgg(fig5, master=plot_frame5)
    canvas5.draw()
    canvas5.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)


    # Plotting the error vs. iteration (optional)
    #plt.figure(figsize=(2, 4))
    #plt.plot(range(1, iteration2 + 1), error3_history, marker='o', linestyle='-', color='b', markersize=5)
    #plt.xlabel('Iteration')
    #plt.ylabel('Error')
    #plt.title('Water Contamination in Oil Layer Iteration')
    #plt.grid(True)
    #plt.tight_layout()
    #plt.show()

    # Plotting the results (optional)
    #plt.figure(figsize=(14, 4))
    #plt.plot(water_layer_thickness_along_pipe[:, 0], water_layer_thickness_along_pipe[:, 1], label='Water Layer Thickness')
    #plt.plot(oil_layer_parameters[:, 0], pipe_diameter_m - oil_layer_parameters[:, 2], label='Oil Layer Thickness')
    #plt.xlabel('Pipe Length (m)')
    #plt.ylabel('Pipe Diameter (m)')
    #plt.ylim(0, pipe_diameter_m)
    #plt.xlim(0, water_layer_thickness_along_pipe[-1, 0])
    #plt.title('Layers Thickness Profile along the Pipe Length')
    #plt.legend()
    #plt.grid(True)
    #plt.show()


    print(f"Oil Layer Thickness: {oil_thickness:.3f} m")
    print(f"Water Layer Thickness: {water_thickness:.3f} m")
    print(f"Emulsion Layer Thickness: {D_in - (water_thickness + oil_thickness):.3f} m")

    # Plotting water fraction vs height
    #plt.figure(figsize=(4, 8))
    #plt.plot(water_fraction, y, label='Water Fraction')  # Swap y and water_fraction
    #plt.plot(cumulative_water_areas, y, label='Cumulative Water Fraction')  # Swap y and cumulative_water_areas
    #plt.plot(widths, y, label='Pipe widths', color='gray')  # Swap y and cumulative_water_areas, set color to gray
    #plt.axvline(initial_water_cut, color='r', linestyle='--', label=f'Total Water Cut = {initial_water_cut:.2f}')  # Use axvline for x-axis line
    #plt.ylabel('Pipe Diameter [m]')  # Swap xlabel and ylabel
    #plt.xlabel('Water Fraction (WC) [-]')  # Swap xlabel and ylabel
    #plt.title('Water Fraction Change from Bottom to Top of the Pipe')
    #plt.legend()
    #plt.grid(True)
    #plt.ylim(0, D_in)  # Set y-axis limits from 0 to D_in
    #plt.show()


    # Plotting water fraction vs width
    #plt.figure(figsize=(8, 4))  # Create a new figure
    #plt.plot(water_fraction, WT, label='Drainage Potential Curve')  # Swap y and water_fraction
    #plt.ylabel('Water Tapped (WT) [%]')  # Swap xlabel and ylabel
    #plt.xlabel('Water cut tapped stream [%]')  # Swap xlabel and ylabel
    #plt.title('Drainage Potential Curve')
    #plt.legend()
    #plt.grid(True)
    #plt.show()

# Function to display image in tkinter window
def display_image():
    try:
        # Get the path to the desktop
        desktop_path = os.path.join(os.path.expanduser("~"), "Desktop")
        
        # Specify the image file name
        image_file = "Figure MPPS.png"  # Assuming it's a PNG file, adjust if different
        
        # Construct the full image path
        image_path = os.path.join(desktop_path, image_file)
        
        # Load image from file
        img = Image.open(image_path)
        
        # Resize the image as needed
        img = img.resize((1200, 200), Image.ANTIALIAS)

        # Convert Image object into Tkinter PhotoImage object
        img_tk = ImageTk.PhotoImage(img)

        # Create a label to display the image
        image_label = tk.Label(root, image=img_tk)
        image_label.image = img_tk  # Keep a reference to the image to prevent garbage collection
        image_label.grid(row=0, column=2, rowspan=7, columnspan=7, padx=10, pady=10, sticky='n')

    except FileNotFoundError:
        tk.messagebox.showerror('Error', 'Image file not found')
    except Exception as e:
        tk.messagebox.showerror('Error', f'Error loading image: {e}')


def width_func(y, d):
    """
    Calculate the width of a pipe at a given height.

    Parameters:
    y (numpy.ndarray): Array of heights from y=0 (lowest point of the pipe)
    d (float): Diameter of the pipe in meters

    Returns:
    numpy.ndarray: Array of widths of the pipe at each height
    """
    r = d / 2
    W = np.zeros_like(y)

    # Ensure y does not exceed the diameter of the pipe
    if np.any(y > d):
        print('Position outside of pipe, y larger than d, diameter of pipe')
        return None

    # Calculate width for y <= r (lower half of the pipe)
    mask_lower = y <= r
    distance_below = r - y[mask_lower]
    W[mask_lower] = 2 * r * np.sin(np.arccos(distance_below / r))

    # Calculate width for y > r (upper half of the pipe)
    mask_upper = y > r
    distance_above = y[mask_upper] - r
    W[mask_upper] = 2 * r * np.sin(np.arccos(distance_above / r))

    # Handle the middle point separately
    if r in y:
        W[y == r] = 2 * r

    return W

def calculate_water_fraction(y, oil_thickness, emulsion_thickness, oil_water_contamination, water_contamination):
    """
    Calculate the water fraction at each height in the pipe.

    Parameters:
    y (numpy.ndarray): Array of heights from y=0 (lowest point of the pipe)
    oil_thickness (float): Thickness of the oil layer in meters
    emulsion_thickness (float): Thickness of the emulsion layer in meters
    oil_water_contamination (float): Water contamination in the oil layer (0 to 1)
    water_contamination (float): Water contamination in the water layer (0 to 1)

    Returns:
    numpy.ndarray: Array of water fractions at each height
    """
    water_fraction = np.zeros_like(y)
    for i in range(len(y)):
        if y[i] <= oil_thickness:
            # Oil layer
            water_fraction[i] = oil_water_contamination
        elif y[i] <= oil_thickness + emulsion_thickness:
            # Emulsion layer (linear change)
            start_emulsion = oil_thickness
            end_emulsion = oil_thickness + emulsion_thickness
            water_fraction[i] = oil_water_contamination + (water_contamination - oil_water_contamination) * ((y[i] - start_emulsion) / (end_emulsion - start_emulsion))
        else:
            # Water layer
            water_fraction[i] = water_contamination

    return water_fraction[::-1]

def calculate_cumulative_water_fraction(y, oil_thickness, emulsion_thickness, oil_water_contamination, water_contamination, d):
    """
    Calculate the cumulative water fraction in the pipe.

    Parameters:
    y (numpy.ndarray): Array of heights from y=0 (lowest point of the pipe)
    oil_thickness (float): Thickness of the oil layer in meters
    emulsion_thickness (float): Thickness of the emulsion layer in meters
    oil_water_contamination (float): Water contamination in the oil layer (0 to 1)
    water_contamination (float): Water contamination in the water layer (0 to 1)
    d (float): Diameter of the pipe in meters

    Returns:
    float: Cumulative water fraction in the pipe
    """
    water_fraction = calculate_water_fraction(y, oil_thickness, emulsion_thickness, oil_water_contamination, water_contamination)
    actual_area = (np.pi / 4) * d**2  # Actual cross-sectional area of the pipe

    cumulative_water_fraction = np.trapz(water_fraction * width_func(y, d), y) / actual_area
    return cumulative_water_fraction

def find_layer_thicknesses(initial_water_cut, emulsion_thickness, oil_water_contamination, water_contamination, d, num_segments=1000, max_iterations=100, tolerance=1e-6):
    """
    Find the oil and water layer thicknesses to match a given initial water cut using Newton-Raphson method.

    Parameters:
    initial_water_cut (float): Initial cumulative water fraction desired (0 to 1)
    emulsion_thickness (float): Thickness of the emulsion layer in meters
    oil_water_contamination (float): Water contamination in the oil layer (0 to 1)
    water_contamination (float): Water contamination in the water layer (0 to 1)
    d (float): Diameter of the pipe in meters
    num_segments (int): Number of segments for discretization
    max_iterations (int): Maximum number of iterations for convergence
    tolerance (float): Tolerance for convergence

    Returns:
    oil_thickness (float): Thickness of the oil layer in meters
    water_thickness (float): Thickness of the water layer in meters
    cumulative_water_fraction (float): Actual cumulative water fraction achieved
    """
    # Initialize oil thickness (initial guess)
    oil_thickness = d * 0.3  # Start with 30% of the pipe diameter

    # Discretize the pipe diameter into segments
    y = np.linspace(0, d, num_segments)

    iteration = 0
    cumulative_water_fraction = 0.0

    while iteration < max_iterations:
        # Calculate current cumulative water fraction
        current_water_fraction = calculate_cumulative_water_fraction(y, oil_thickness, emulsion_thickness, oil_water_contamination, water_contamination, d)

        # Calculate derivative of water fraction with respect to oil thickness using central difference
        h = 1e-6
        water_fraction_plus = calculate_cumulative_water_fraction(y, oil_thickness + h, emulsion_thickness, oil_water_contamination, water_contamination, d)
        water_fraction_minus = calculate_cumulative_water_fraction(y, oil_thickness - h, emulsion_thickness, oil_water_contamination, water_contamination, d)
        derivative = (water_fraction_plus - water_fraction_minus) / (2 * h)

        # Update oil thickness using Newton-Raphson method
        oil_thickness -= (current_water_fraction - initial_water_cut) / derivative

        # Recalculate cumulative water fraction
        cumulative_water_fraction = calculate_cumulative_water_fraction(y, oil_thickness, emulsion_thickness, oil_water_contamination, water_contamination, d)

        # Check convergence
        if abs(current_water_fraction - initial_water_cut) < tolerance:
            break

        iteration += 1

    if iteration == max_iterations:
        print(f"Maximum iterations ({max_iterations}) reached without convergence for oil and water thickness.")

    water_thickness = d - oil_thickness - emulsion_thickness

    return oil_thickness, water_thickness, cumulative_water_fraction

def calculate_areas_and_water_fraction(D_in, y, widths, water_fraction, initial_water_cut):
    """
    Calculate the cumulative area of each segment, the cumulative water area of each segment,
    and the normalized cumulative water area.

    Parameters:
    D_in (float): Inner diameter of the pipe in meters.
    y (numpy.ndarray): Array of heights from y=0 (lowest point of the pipe).
    widths (numpy.ndarray): Array of widths of the pipe at each height.
    water_fraction (numpy.ndarray): Array of water fractions at each height.
    initial_water_cut (float): Initial cumulative water fraction desired (0 to 1).

    Returns:
    tuple: cumulative_areas (numpy.ndarray), cumulative_water_areas (numpy.ndarray), WT (numpy.ndarray)
    """
    # Calculate the actual area of the pipe
    actual_area = (np.pi / 4) * D_in**2

    # Calculate the cumulative area of each segment
    cumulative_areas = np.zeros_like(y)
    for i in range(1, len(y)):
        segment_area = (y[i] - y[i - 1]) * ((widths[i] + widths[i - 1]) / 2)
        cumulative_areas[i] = cumulative_areas[i - 1] + segment_area

    # Calculate the cumulative water area of each segment
    cumulative_water_areas = np.zeros_like(y)
    for i in range(1, len(y)):
        segment_water_area = (y[i] - y[i - 1]) * ((widths[i] + widths[i - 1]) / 2) * ((water_fraction[i] + water_fraction[i - 1]) / 2) / actual_area
        cumulative_water_areas[i] = cumulative_water_areas[i - 1] + segment_water_area

    # Normalize cumulative water areas by the initial water cut
    WT = cumulative_water_areas / initial_water_cut

    return cumulative_areas, cumulative_water_areas, WT


def calculate_flow_parameters(flow_rate_lpm, pipe_diameter_m, number_branches):
    flow_rate_m3s = flow_rate_lpm / 1000 / 60 / number_branches
    pipe_radius_m = pipe_diameter_m / 2
    pipe_area_m2 = math.pi * (pipe_radius_m ** 2)
    mixture_velocity_mps = flow_rate_m3s / pipe_area_m2

    return {
        'pipe_area_m2': pipe_area_m2,
        'mixture_velocity_mps': mixture_velocity_mps,
        'flow_rate_m3s': flow_rate_m3s
    }


def calculate_slip_velocity(water_density, water_viscosity, water_fraction, oil_density, oil_viscosity, droplet_diameter, inversion_point):
    emulsion_type = "OiW" if water_fraction > inversion_point else "WiO"
    if (water_fraction > inversion_point):
        continuous_density = water_density
        continuous_viscosity = water_viscosity
        continuous_fraction = water_fraction
        dispersed_density = oil_density
        dispersed_viscosity = oil_viscosity
        dispersed_fraction = 1 - water_fraction
    else:
        continuous_density = oil_density
        continuous_viscosity = oil_viscosity
        continuous_fraction = 1 - water_fraction
        dispersed_density = water_density
        dispersed_viscosity = water_viscosity
        dispersed_fraction = water_fraction

    hr_factor = (3 * (continuous_viscosity + dispersed_viscosity)) / ((2 * continuous_viscosity) + (3 * dispersed_viscosity))
    svp1 = 5 * (hr_factor**(-3/2)) * ((dispersed_fraction / continuous_fraction)**0.45)
    svp2 = (continuous_fraction / (2 * dispersed_fraction * hr_factor)) * math.exp((2.5 * dispersed_fraction) / (1 - (0.61 * dispersed_fraction)))
    g = 9.81
    arch_num = (continuous_density * abs(continuous_density - dispersed_density) * g * (droplet_diameter**3)) / (continuous_viscosity**2)
    reynolds_num = 9.72 * ((1 + ((0.01 * arch_num)**(4/7))) - 1)
    friction_coeff = (arch_num / (6 * (reynolds_num**2))) - (3 / (hr_factor * reynolds_num))
    term1 = (3 * svp2 * dispersed_fraction * continuous_viscosity) / (friction_coeff * svp1 * continuous_fraction * continuous_density * droplet_diameter)
    term2 = (1 + ((arch_num * friction_coeff * svp1 * (continuous_fraction**3)) / (54 * (svp2**2) * (dispersed_fraction**2))))**0.5 - 1
    slip_velocity_value = term1 * term2

    return slip_velocity_value, continuous_viscosity


def calculate_water_layer_thickness(slip_velocity, pipe_diameter, mixture_velocity, water_fraction, step_length=0.01):
    def calculate_layer_thickness(pipe_diameter, water_fraction):
        pipe_radius = pipe_diameter / 2
        pipe_area = math.pi * pipe_radius**2

        if water_fraction <= 0.5:
            area_water = water_fraction * pipe_area
            def f(layer_thickness):
                return (pipe_radius**2) * math.acos((pipe_radius - layer_thickness) / pipe_radius) \
                       - (pipe_radius - layer_thickness) * ((2 * pipe_radius * layer_thickness - layer_thickness**2)**0.5)
            lower_bound = 0.0
            upper_bound = pipe_radius
            tolerance = 1e-6

            while upper_bound - lower_bound > tolerance:
                mid = (lower_bound + upper_bound) / 2.0
                if f(mid) < area_water:
                    lower_bound = mid
                else:
                    upper_bound = mid

            water_layer_thickness_final = (lower_bound + upper_bound) / 2.0
            oil_layer_thickness_final = pipe_radius - water_layer_thickness_final

        else:
            oil_fraction = 1 - water_fraction
            area_oil = oil_fraction * pipe_area
            def g(layer_thickness):
                return (pipe_radius**2) * math.acos((pipe_radius - layer_thickness) / pipe_radius) \
                       - (pipe_radius - layer_thickness) * ((2 * pipe_radius * layer_thickness - layer_thickness**2)**0.5)
            lower_bound = 0.0
            upper_bound = pipe_radius
            tolerance = 1e-6

            while upper_bound - lower_bound > tolerance:
                mid = (lower_bound + upper_bound) / 2.0
                if g(mid) < area_oil:
                    lower_bound = mid
                else:
                    upper_bound = mid

            oil_layer_thickness_final = (lower_bound + upper_bound) / 2.0
            water_layer_thickness_final = 2 * pipe_radius - oil_layer_thickness_final

        return water_layer_thickness_final, oil_layer_thickness_final

    water_layer_thickness_final, oil_layer_thickness_final = calculate_layer_thickness(pipe_diameter, water_fraction)
    current_length = 0
    current_water_layer_thickness = 0
    results = []

    while current_water_layer_thickness < water_layer_thickness_final:
        results.append([current_length, current_water_layer_thickness])
        current_length += step_length
        current_water_layer_thickness = slip_velocity * current_length / mixture_velocity

    results.append([current_length, water_layer_thickness_final])
    results = np.array(results)

    return results, water_layer_thickness_final


def calculate_oil_layer_parameters(water_layer_thickness_along_pipe, water_density, oil_density, interfacial_tension, continuous_phase_viscosity, asymmetry_parameter, mixture_velocity, initial_droplet_diameter=0.001, initial_oil_layer_thickness=0.000001, hamaker_coefficient=1e-20, step_length=0.01):
    num_steps = len(water_layer_thickness_along_pipe)
    results = np.zeros((num_steps, 11))
    results[:, 0] = water_layer_thickness_along_pipe[:, 0]
    results[0, 1] = initial_droplet_diameter
    results[0, 2] = initial_oil_layer_thickness
    g = 9.81
    pi = math.pi

    for i in range(num_steps):
        if i > 0:
            results[i, 1] = results[i-1, 1] + results[i-1, 9]
            results[i, 2] = results[i-1, 2] + results[i-1, 10]

        results[i, 3] = ((abs(water_density - oil_density) * g / interfacial_tension)**0.6) * (results[i, 2]**0.2) * results[i, 1]
        results[i, 4] = 0.5 * results[i, 1] * (1 - (math.sqrt(1 - (4.7 / (results[i, 3] + 4.7)))))
        results[i, 5] = 0.3025 * results[i, 1] * (math.sqrt(1 - (4.7 / (results[i, 3] + 4.7))))
        results[i, 6] = math.sqrt(3) * results[i, 5]
        results[i, 7] = (((6 * pi)**(7/6)) * continuous_phase_viscosity * (results[i, 4]**(7/3))) / (4 * (interfacial_tension**(5/6)) * (hamaker_coefficient**(1/6)) * results[i, 5] * asymmetry_parameter)
        results[i, 8] = (((6 * pi)**(7/6)) * continuous_phase_viscosity * (results[i, 4]**(7/3))) / (4 * (interfacial_tension**(5/6)) * (hamaker_coefficient**(1/6)) * results[i, 6] * asymmetry_parameter)
        results[i, 9] = (results[i, 1] * step_length) / (mixture_velocity * 6 * results[i, 7])
        results[i, 10] = (step_length * 2 * results[i, 1]) / (mixture_velocity * 3 * results[i, 8])

    return results

# Display the image when the GUI is initialized
display_image()

# Create a button for other purposes
action_button = tk.Button(root, text='Action Button', command=on_button_click)
action_button.grid(row=8, column = 2, columnspan=2, pady=10)

# Run the tkinter main loop
root.mainloop()
```
