```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt

x_axis_labels = [30,50,70,90]
y_axis_labels = [700,500,300]
sb.set(rc = {'figure.figsize':(15,12)})
sb.set(font_scale= 2)



ax = sb.heatmap(SE1, cmap='RdYlGn', vmin=30, vmax=100, annot=True, cbar_kws={'label': 'My Colorbar', 'orientation': 'horizontal'}, cbar=False, xticklabels=x_axis_labels, yticklabels=y_axis_labels, fmt='.1f', linewidths=0.5, linecolor='white')

plt.xlabel("Water cut [%]", fontsize=39,fontweight='bold')
plt.ylabel("Total flow rate [L/min]", fontsize=38,fontweight='bold')

ax.axhline(y=0, color='k',linewidth=5)
ax.axhline(y=9, color='k',linewidth=5)
ax.axvline(x=0, color='k',linewidth=5)
ax.axvline(x=4, color='k',linewidth=5)

ax.set_xticklabels(ax.get_xticklabels(), fontsize = 29,fontweight='bold')
ax.set_yticklabels(ax.get_yticklabels(), fontsize = 29, rotation=0,fontweight='bold')

#adjust y-axis label position
ax.yaxis.set_label_coords(-.1, .5)

#adjust x-axis label position
ax.xaxis.set_label_coords(.5, -.1)

#adjust axis tick
ax.tick_params(left=True, bottom=True, size=8, width=5)
plt.show()
```
